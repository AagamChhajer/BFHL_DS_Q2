```python
import cv2
import pytesseract
import numpy as np
from PIL import Image
import re
def preprocess_image(image_path):
    # Read image
    img = cv2.imread(image_path)
    
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Resize image
    resized = cv2.resize(gray, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
    
    # Apply adaptive thresholding
    processed = cv2.adaptiveThreshold(
        resized,
        255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY,
        61,
        11
    )
    
    return processed

def extract_lab_data(image_path):
    # Set Tesseract path
    pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
    
    # Preprocess image
    # processed_img = preprocess_image(image_path)
    processed_img = cv2.imread(image_path)
    
    # Extract text using OCR
    text = pytesseract.image_to_string(processed_img)
    print("Extracted Text:",text)
    # Define patterns for lab data
    patterns = {
        'test_name': r'([A-Za-z\s]+):\s*',
        'value': r':\s*([\d.]+)\s*',
        'reference_range': r'\(([\d.-]+)\)'
    }
    
    # Extract lab test data
    results = []
    lines = text.split('\n')
    
    for line in lines:
        if ':' in line:  # Potential test result line
            result = {}
            
            # Extract test name
            test_match = re.search(patterns['test_name'], line)
            if test_match:
                result['test_name'] = test_match.group(1).strip()
                
                # Extract value
                value_match = re.search(patterns['value'], line)
                if value_match:
                    result['value'] = value_match.group(1)
                    
                    # Extract reference range
                    range_match = re.search(patterns['reference_range'], line)
                    result['reference_range'] = range_match.group(1) if range_match else "N/A"
                    
                    results.append(result)
    
    return results

def main():
    # Replace with your image path
    image_path = r"C:\Users\HP\Downloads\bajaj_ds\bajaj_q2\dataset\IMG-20250429-WA0009.jpg"
    
    try:
        # Process image and extract data
        results = extract_lab_data(image_path)
        
        # Print results
        print("\nExtracted Lab Test Results:")
        print("-" * 50)
        for result in results:
            print(f"Test Name: {result['test_name']}")
            print(f"Value: {result['value']}")
            print(f"Reference Range: {result['reference_range']}")
            print("-" * 50)
            
    except Exception as e:
        print(f"Error processing image: {str(e)}")

if __name__ == "__main__":
    main()
```
